```python
def make_graph(r1, r2, r3, r4, step=1):
    fig = plt.figure(figsize=(8,6))

    h = plt.hist(r1[::1], bins='fd', linewidth=1.5, histtype='step', color=colors[0])
    h = plt.hist(r1[::1], bins='fd', linewidth=0, alpha=0.2, color=colors[0])

    h = plt.hist(r2[::1], bins='fd', linewidth=1.5, linestyle='-', histtype='step', color=colors[1])
    h = plt.hist(r2[::1], bins='fd', linewidth=0, linestyle='-', alpha=0.2, color=colors[1])

    h = plt.hist(r3[::1], bins='fd', linewidth=2, histtype='step', color=colors[2])

    h = plt.hist(r4[::1], bins='fd', linewidth=2, histtype='step', color=colors[3])
    plt.xlim(0.5, 2)
    
    return fig
```

```python
def prune(arr):
    arr = arr.flatten()
    arr[np.isinf(arr)] = 10
    return arr[~np.isnan(arr)]
```

```python
for ph in [500]:
    for cond in ['storm_inner']:
        r_m_inner = prune(np.load(os.path.join(data_dir, 'r_values', 'r_inner_m_ph_{}_{}.npy'.format(ph, cond))))
        r_m_outer = prune(np.load(os.path.join(data_dir, 'r_values', 'r_outer_m_ph_{}_{}.npy'.format(ph, cond))))
        r_gt_inner = prune(np.load(os.path.join(data_dir, 'r_values', 'r_inner_gt_ph_{}_{}.npy'.format(ph, cond))))
        r_gt_outer = prune(np.load(os.path.join(data_dir, 'r_values', 'r_outer_gt_ph_{}_{}.npy'.format(ph, cond))))
        
        fig = make_graph(r_gt_inner, r_gt_outer, r_m_inner, r_m_outer)
        
        plt.savefig('hist_{}_{}.png'.format(ph, cond), dpi=600)
```
